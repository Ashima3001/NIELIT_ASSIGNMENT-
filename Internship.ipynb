```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import (
    MinMaxScaler,
        StandardScaler,
            MaxAbsScaler,
                RobustScaler,
                    Normalizer
                    )
```

```python
data = df[['LotArea', 'GrLivArea', 'TotalBsmtSF', 'SalePrice']]

print(data.head())
```

```python
minmax = MinMaxScaler()

minmax_data = pd.DataFrame(
    minmax.fit_transform(data),
        columns=data.columns
        )
print(minmax_data.head())
```

```python
standard = StandardScaler()

standard_data = pd.DataFrame(
    standard.fit_transform(data),
        columns=data.columns
        )

print(standard_data.head())
```

```python
maxabs = MaxAbsScaler()

maxabs_data = pd.DataFrame(
    maxabs.fit_transform(data),
        columns=data.columns
        )

print(maxabs_data.head())
```

```python
log_data = np.log1p(data)

print(log_data.head())
```

```python
normalizer = Normalizer()

normalized_data = pd.DataFrame(
    normalizer.fit_transform(data),
        columns=data.columns
        )
print(normalized_data.head())
```

A. Which scaler converts values between 0 and 1?
Answer: Min-Max Scaler
B. Which scaler is most suitable when outliers are present?
Answer: Robust Scaler
C. Which normalization converts every sample into unit length?
Answer: Unit Vector Normalization (Normalizer)
